```python
import os
import cv2 #computer vision
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# download dataset directly from tensorflow
mnist = tf.keras.datasets.mnist
```

```python
# training data - used to train the model
# testing data - used to assess the model
# x data is the image itself
# y data is the classification, number/digit

#call load data function, returning 2 tuples with training and testing data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# normalizing data, i.e. scaling it down, such that each value is between 0 and 1
# normalize the pixels only (easier for neural network to do calcs)
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
# neural network
# create model
model = tf.keras.models.Sequential()

# add a layer
# flatten input shapes
model.add(tf.keras.layers.Flatten(input_shape = (28,28)))
# dense layer
# relu = rectify linear unit
model.add(tf.keras.layers.Dense(128, activation='relu'))
# softmax - give probability of digit to be the right answer
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
model.save('handwritten.model.keras')
```

```python
model = tf.keras.models.load_model('handwritten.model.keras')
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(loss)
print(accuracy)

# low loss, high accuracy
```

```python
image_number = 1
while os.path.isfile(f"digits/digit{image_number}.png"):
    try:
        img = cv2.imread(f"digits/digit{image_number}.png")[:,:,0]
        img = np.invert(np.array([img]))
        prediction = model.predict(img)
        print(f"This digit is probably a {np.argmax(prediction)}")
        plt.imshow(img[0], cmap=plt.cm.binary)
        plt.show()
    except:
        print("Error!")
    finally:
        image_number += 1
```
